# Get all the sequence files

```python
conn = Connection("prod", True)
```

```python
def get_sequence_file_info(sequence_file):
    return {
        "accession": sequence_file["accession"],
        "uuid": sequence_file["uuid"],
        "href": sequence_file["href"],
        "md5sum": sequence_file["md5sum"],
        "aliases": sequence_file["aliases"],
        "file_set": sequence_file["file_set"]["accession"],
        "summary": sequence_file['file_set']["summary"],
        "file_size": sequence_file["file_size"],
        "lane": sequence_file["lane"],
        "flowcell_id": sequence_file["flowcell_id"],
        "sequencing_kit": sequence_file["sequencing_kit"],
        "sequencing_run": sequence_file["sequencing_run"],
        "sequencing_platform": sequence_file["sequencing_platform"]['term_name'],
        "illumina_read_type": sequence_file["illumina_read_type"],
        "minimum_read_length": sequence_file["minimum_read_length"],
        "maximum_read_length": sequence_file["maximum_read_length"],
        "submitted_file_name": sequence_file["submitted_file_name"],
    }
```

```python
# Go one by one through the list of files and get the information
scRNA_file_info = []
sgRNA_file_info = []
for idx, item_id in tqdm.tqdm(enumerate(df['@id']), total=len(df), desc="Getting sequence file info"):
    accession = item_id.split('/')[-2]
    print(accession)
    measurement_set = conn.get(f"/measurement_set/{accession}")
    for sequence_file in measurement_set["files"]:
        sequence_file_accession = sequence_file["accession"]
        sequence_file_conn = conn.get(f"/sequence_file/{sequence_file_accession}")
        if sequence_file_conn is not None:
            scRNA_file_info.append(get_sequence_file_info(sequence_file_conn))

    auxiliary_set = conn.get(measurement_set["auxiliary_sets"][0]["@id"])
    for auxiliary_file in auxiliary_set["files"]:
        auxiliary_file_accession = auxiliary_file["accession"]
        auxiliary_file_conn = conn.get(f"/sequence_file/{auxiliary_file_accession}")
        if auxiliary_file_conn is not None:
            auxiliary_file_dict = get_sequence_file_info(auxiliary_file_conn)
            auxiliary_file_dict["measurement_set"] = measurement_set["@id"].split("/")[-2]
            sgRNA_file_info.append(auxiliary_file_dict)

scRNA_df = pd.DataFrame(scRNA_file_info)
sgRNA_df = pd.DataFrame(sgRNA_file_info)
```

# Double check fastqs

```python
FASTQ_DIR = "/cellar/users/aklie/data/datasets/tf_perturb_seq/datasets/Huangfu_HUES8-embryonic-stemcell-differentiation_TF-Perturb-seq/fastq_files"
```

```python
import hashlib
import glob

def md5_file(filepath, chunk_size=8192):
    md5 = hashlib.md5()
    with open(filepath, 'rb') as f:
        while True:
            chunk = f.read(chunk_size)
            if not chunk:
                break
            md5.update(chunk)
    return md5.hexdigest()
```

```python
md5sum_dict = scRNA_df_filt.set_index("accession")["md5sum"]
for idx, row in sgRNA_df_filt.iterrows():
    accession = row["accession"]
    md5sum_dict[accession] = row["md5sum"]
len(md5sum_dict)
```

```python
# for each file in FASTQ_DIR
local_md5sum_dict = {}
bad_accessions = []
for fastq_file in tqdm.tqdm(glob.glob(f"{FASTQ_DIR}/*.fastq.gz"), desc="Calculating local md5sums"):
    accession = fastq_file.split("/")[-1].split(".")[0]
    md5sum = md5sum_dict.get(accession)
    local_md5sum = md5_file(fastq_file)
    if local_md5sum != md5sum:
        print(f"MD5 mismatch for {accession}: {local_md5sum} != {md5sum}")
        bad_accessions.append(accession)
    local_md5sum_dict[accession] = local_md5sum
```

```python
# Save local md5sum to csv
local_md5sum_df = pd.DataFrame.from_dict(local_md5sum_dict, orient='index', columns=['local_md5sum'])
local_md5sum_df.index.name = 'accession'
local_md5sum_df.to_csv(os.path.join(FASTQ_DIR, "md5sums.tsv"), sep="\t")
```

# Make samplesheet

```python
scRNA_R1_df = scRNA_df_filt[scRNA_df_filt["illumina_read_type"] == "R1"]
scRNA_R2_df = scRNA_df_filt[scRNA_df_filt["illumina_read_type"] == "R2"]

scRNA_merged = pd.merge(
    scRNA_R1_df,
    scRNA_R2_df,
    on=["file_set", "lane", "flowcell_id"],
    suffixes=("_r1", "_r2"),
)

scRNA_samplesheet = pd.DataFrame({
    "R1_path": FASTQ_DIR + "/" + scRNA_merged["accession_r1"] + ".fastq.gz",
    "R2_path": FASTQ_DIR + "/" + scRNA_merged["accession_r2"] + ".fastq.gz",
    "measurement_sets": scRNA_merged["file_set"],
    "Lane": scRNA_merged["lane"].astype(str) + "_" + scRNA_merged["flowcell_id"],
    "file_modality": "scRNA"
})

scRNA_samplesheet = scRNA_samplesheet.sort_values(by=["Lane"])
scRNA_samplesheet
```
