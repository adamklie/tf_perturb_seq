# TF Enrichment Comparison

Compare ChIP-seq TF enrichment analysis results across datasets.

Uses the pipeline's `tf/benchmark_output/benchmark_tables/` output which tests whether 
DEGs of each perturbed TF are enriched for that TF's ChIP-seq binding targets.

**Input:** TF benchmark tables from `tf/benchmark_output/benchmark_tables/` per dataset  
**Output:** Cross-dataset comparison of TF enrichment

```python
import os
import sys
from pathlib import Path

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
```

```python
# Project root
PROJECT_ROOT = Path("/cellar/users/aklie/projects/tf_perturb_seq")

sys.path.append(str(PROJECT_ROOT / "config"))
from loader import load_colors

dataset_colors = load_colors("technology-benchmark_WTC11_TF-Perturb-seq", "dataset_colors")
dataset_order = load_colors("technology-benchmark_WTC11_TF-Perturb-seq", "dataset_order")
```

```python
# Parameters
RUN_LABEL = "cleanser_unified"
```

```python
# Derived paths
BASE_DIR = PROJECT_ROOT / "datasets" / "technology-benchmark_WTC11_TF-Perturb-seq"
OUTPUT_DIR = BASE_DIR / "results" / "cross_tech_comparison" / RUN_LABEL / "inference"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
DATASETS_ROOT = PROJECT_ROOT / "datasets"

DATASETS = [
    "Hon_WTC11-benchmark_TF-Perturb-seq",
    "Huangfu_WTC11-benchmark_TF-Perturb-seq",
    "Gersbach_WTC11-benchmark_TF-Perturb-seq_GEM-Xv3",
    "Gersbach_WTC11-benchmark_TF-Perturb-seq_HTv2",
    "Engreitz_WTC11-benchmark_TF-Perturb-seq",
]

SUFFIX = "_WTC11-benchmark_TF-Perturb-seq"
def short_name(ds):
    return ds.replace(SUFFIX, "").replace("_", " ")
```

```python
# Load TF enrichment tables
enrichment_dfs = {}
tf_peak_dfs = {}

for ds in DATASETS:
    tf_dir = DATASETS_ROOT / ds / "runs" / RUN_LABEL / "tf" / "benchmark_output" / "benchmark_tables"
    
    enrich_path = tf_dir / "enrichment_all.tsv"
    peak_path = tf_dir / "tf_peak_mapping.tsv"
    
    if enrich_path.exists():
        enrichment_dfs[ds] = pd.read_csv(enrich_path, sep="\t")
        enrichment_dfs[ds]["dataset"] = ds
        print(f"  {short_name(ds)}: {len(enrichment_dfs[ds])} enrichment rows")
    else:
        print(f"  {short_name(ds)}: MISSING enrichment_all.tsv at {tf_dir}")
    
    if peak_path.exists():
        tf_peak_dfs[ds] = pd.read_csv(peak_path, sep="\t")

print(f"\nLoaded {len(enrichment_dfs)} enrichment tables, {len(tf_peak_dfs)} peak mappings")
```
