```python
import os
import glob
```

```python
run_name = "2026_01_14"
```

```python
datasets_root = "/cellar/users/aklie/data/datasets/tf_perturb_seq/datasets"
all_datasets = sorted(glob.glob(os.path.join(datasets_root, "*")))
all_datasets
```

```python
# Filter for "benchmark"
benchmark_datasets = [d for d in all_datasets if "benchmark" in d][:-1]
benchmark_datasets
```

```python
inputs = [
    "/cellar/users/aklie/data/datasets/tf_perturb_seq/datasets/Gersbach_WTC11-benchmark_TF-Perturb-seq_GEM-Xv3/results/1_CRISPR_pipeline/2026_01_06/inference_mudata.h5mu",
    "/cellar/users/aklie/data/datasets/tf_perturb_seq/datasets/Gersbach_WTC11-benchmark_TF-Perturb-seq_HTv2/results/1_CRISPR_pipeline/2026_01_06/inference_mudata.h5mu",
    "/cellar/users/aklie/data/datasets/tf_perturb_seq/datasets/Hon_WTC11-benchmark_TF-Perturb-seq/results/1_CRISPR_pipeline/2026_01_05/inference_mudata.h5mu",
    "/cellar/users/aklie/data/datasets/tf_perturb_seq/datasets/Huangfu_WTC11-benchmark_TF-Perturb-seq/results/1_CRISPR_pipeline/2026_01_12/pipeline_dashboard/inference_mudata.h5mu"
]
```

```python
# Make outdirs as /2_qc extension of dataset name
outdirs = [datasets + f"/results/2_qc/{run_name}" for datasets in benchmark_datasets]
outdirs
```

```python
dataset_names = [os.path.basename(d) for d in benchmark_datasets]
dataset_names
```

```python
# TSV format (header required):
#   input<TAB>outdir<TAB>run_name
#   /path/A.h5mu<TAB>/out/A<TAB>A
#   /path/B.h5mu<TAB>/out/B<TAB>B
with open("/cellar/users/aklie/data/datasets/tf_perturb_seq/datasets/qc.tsv", "w") as f:
    f.write("input\toutdir\trun_name\n")
    for inp, outdir, name in zip(inputs, outdirs, dataset_names):
        f.write(f"{inp}\t{outdir}\t{name}\n")
```
