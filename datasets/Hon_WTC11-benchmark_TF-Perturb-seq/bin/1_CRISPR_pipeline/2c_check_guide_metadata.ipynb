# Filter and save

```python
# Remove all rows with duplicate guide_ids
guide_metadata_filt = guide_metadata[~guide_metadata['guide_id'].duplicated(keep=False)]
print("Number of rows after removing duplicates:", guide_metadata_filt.shape[0])
```

```python
# Remove all rows with type == "positive control"
guide_metadata_filt = guide_metadata_filt[guide_metadata_filt["type"] != "positive control"]
print("Number of rows after removing positive controls:", guide_metadata_filt.shape[0])
```

```python
# Remove all with intended_target_name is NA
guide_metadata_filt = guide_metadata_filt[~guide_metadata_filt["intended_target_name"].isna()]
print("Number of rows after removing NA intended_target_name:", guide_metadata_filt.shape[0])
```

```python
guide_metadata_filt["type"].value_counts()
```

```python
guide_metadata_filt[guide_metadata_filt["intended_target_chr"] == "chrPC"]
```

```python
# Save
path_guide_metadata_filt = path_guide_metadata.replace(".tsv", "_filt.tsv")
guide_metadata_filt.to_csv(path_guide_metadata_filt, sep="\t", index=False)
print("Saved filtered guide metadata to:", path_guide_metadata_filt)
```
